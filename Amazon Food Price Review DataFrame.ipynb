```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
import re

product_id = []
user_id = []
profile_name = []
helpfulness_numerator = []
helpfulness_denominator = []
score = []
time = []
review_summary = []
review_text = []


for i in range(1,568454):
    
    f = open(r'C:\Users\gaura\OneDrive\Desktop\nlp project\txt_reviews\review_'+str(i)+'.txt', "r",encoding="utf8")
    b = f.read().split('\n')

    product_id.append(b[0].split(':')[1].strip(' '))
    user_id.append(b[1].split(':')[1].strip(' '))
    profile_name.append(b[2].split(':')[1].strip(' '))
    helpfulness_numerator.append(b[3].split(':')[1].strip(' '))
    helpfulness_denominator.append(b[4].split(':')[1].strip(' '))
    score.append(b[5].split(':')[1].strip(' '))
    time.append(b[6].split(':')[1].strip(' '))
    review_summary.append(b[7].split(':')[1].strip(' '))
    review_text.append(b[8].split(':')[1].strip(' '))

```

```python
data = pd.DataFrame()
```

```python
data = pd.DataFrame()

data['product_id'] = product_id
data['user_id']=user_id
data['profile_name']=profile_name
data['helpfulness_numerator'] = helpfulness_numerator
data['helpfulness_denominator']=helpfulness_denominator
data['rating']=score
data['time']=time
data['review_summary']=review_summary
data['review_text']=review_text
```

```python
#helpfulness

data['helpfulness_numerator'] = data['helpfulness_numerator'].apply(lambda x :int(x))

data['helpfulness_denominator'] = data['helpfulness_denominator'].apply(lambda x : int(x))
```

```python
data['helpfulness'] = data['helpfulness_numerator'].apply(lambda x :x)/data['helpfulness_denominator'].apply(lambda x : x)
```

```python
data['rating'] = data['rating'].apply(lambda x : int(x))
```

```python
data['sentiment'] = data['rating'].apply(lambda x : 'positive' if x>3 else('negative' if x<3 else 'not defined')) 
```

```python
data["usefulness"] = data['helpfulness'].apply(lambda n: ">75%" if n > 0.75 else ("<25%" if n < 0.25 else 
                                                                                  ("25-75%" if n >= 0.25 and
                                                                        n <= 0.75 else "useless")))
```

```python
data = data.dropna(how = 'any')
```
